```python
f =open('파이썬데이터분석-강의자료/data/csv/weather.csv')
data = csv.reader(f)
header = next(data)

max_wind = 0.0

for row in data:
    if row[2] == '':
        wind = 0
    else:
        wind = float(row[2])
    if max_wind < wind:
        max_wind = wind

print("지난 10년간 울릉도의 최대 풍속은 ", max_wind, "m/s")
```

```python
# 울릉도는 몇 월에 바람이 가장 강핳까?

import csv
import matplotlib.pyplot as plt

f = open('파이썬데이터분석-강의자료/data/csv/weather.csv')
data = csv.reader(f)
header = next(data)

monthly_wind = [0 for x in range(12)]   # 매달 풍속을 담을 리수트, 초기화 0
days_counted = [0 for x in range(12)]   # 각 달마다 측정한 실수, 초기화 0

for row in data:
    month = int(row[0][5:7])    #  0번 열에서 달 정보 추출
    if row[3] != '':    # 풍속 데이터가 존재하는지 확인
        wind = float(row[3])    # 풍속을 얻어 온다
        monthly_wind[month-1] += wind   # 해당 달에 풍속 데이터 추가
        days_counted[month-1] += 1      # 해당 달의 일수를 증가
    
for i in range(12):
    monthly_wind[i] /= days_counted[1]  # 일수로 나누어 월평균 구하기

plt.plot(monthly_wind, 'b-')
plt.show()

f.close()
```

```python
# 윤곽선 뽑아내기

img_gray = cv2.imread('python_analize/data/image/green_back.png', cv2.IMREAD_GRAYSCALE)

# adaptivThreshold를 적용한다
# 주변 9x9 픽셀 공간의 평균값 -5가  임계치가 되고, 이보다 크면 255, 그렇지 않으면 0
img_edge = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                  cv2.THRESH_BINARY, blockSize=9, C=5)

#결과는 윤곽선 픽셀이 드러난다.
cv2.imshow('egde', img_edge)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 합성 사진 만들기2

import numpy as np
import cv2

img1 = cv2.imread('python_analize/data/image/green_back.png')
img2 = cv2.imread('python_analize/data/image/iceberg.png')

front_image = cv2.resize(img1, (300,400))
back_image = cv2.resize(img2, (300, 400))

img_hsv = cv2.cvtColor(front_image, cv2.COLOR_BGR2HSV)
l_bound = np.array([40, 100, 50])
u_bound = np.array([80, 255, 255])

my_mask = cv2.inRange(img_hsv, l_bound, u_bound)
mask_inv = cv2.bitwise_not(my_mask)

# 녹색 픽셀들만 추출하기
extracted = cv2.bitwise_and(front_image, front_image, mask=my_mask)
#녹색 아닌 픽셀만 추출하기
removed = cv2.bitwise_and(front_image, front_image, mask=mask_inv)
# 녹색과 겹치는 배경 추출하기
background = cv2.bitwise_and(back_image, back_image, mask=my_mask)
# 녹색 제거 전경 + 배경 추출하기
merged = cv2.bitwise_or(removed, background)

cv2.imshow('mask', my_mask)
cv2.imshow('mask_inv', mask_inv)
cv2.imshow('removed', removed)
cv2.imshow('background', background)
cv2.imshow('merged', merged)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
