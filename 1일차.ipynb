```python
import string
src_str = string.ascii_uppercase
print(f'src_str = {src_str}')

dst_str = src_str[1:] + src_str[:1]
print(f'dst_str = {dst_str}')
```

```python
upper_src_str = string.ascii_uppercase
upper_dst_str = upper_src_str[3:] + upper_src_str[:3]

lower_src_str = string.ascii_lowercase
lower_dst_str = lower_src_str[3:] + lower_src_str[:3]


def cipher(a):
    if a in upper_src_str:
        idx = upper_src_str.index(a)
        return upper_dst_str[idx]
    else:
        idx = lower_src_str.index(a)
        return lower_dst_str[idx]

src = input("문장을 입력하시오: ")
print("암호화된 문장: ", end='')

for ch in src:
    if ch in upper_src_str or ch in lower_src_str:
        print(cipher(ch), end='')
    else:
        print(ch, end='')
print()
```

```python
import wikipedia

wiki = wikipedia.page('Elon Reeve Musk')
text = wiki.content

from wordcloud import WordCloud, STOPWORDS

s_words = STOPWORDS.union({'one', 'using', 'first', 'two', 'make', 'use', 'made' , 'share', 'said'})
wordcloud = WordCloud(width=2000, height=1500, stopwords=s_words).generate(text)

import matplotlib.pyplot as plt

plt.figure(figsize=(40, 30))
plt.imshow(wordcloud)
plt.show()
```

```python
import re

txt1 = "Life is too short, you need python."
print(re.search('Life', txt1))
txt2 = "The best monents of my life."
print(re.search('Life', txt2))  # 대소문자 구분됨
match = re.search('Life', txt1)
print(match.group())
print(match.start())
print(match.end())
print(match.span())
print(re.search('Life|life', txt2))
print(re.search('[Ll]ife', txt2))   # 문장 안에 Life 혹은 life가 있는지 , []는 문자 선택의 범위 표시
```

```python
txt3 = "My Life My Choice"
print(re.search('^Life', txt1))
print(re.search('^Life', txt2))
print(re.search('^Life', txt3), '\n')     # 메타문자 ^에 의하여 문장의 첫 단어가 Life인 문장만 검색

txt1 = 'Who are you to judge the life I live'
txt2 = 'The best moments of my life'
print(re.search('life$', txt1))
print(re.search('life$', txt2))
```
